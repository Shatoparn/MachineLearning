```python
X = np.array([
    [-2, 4, 1],
    [4, 1, 1],
    [1, 6, 1],
    [2, 4, 1],
    [6, 2, 1]
])
```

```python
def sum_sgd(X, Y):
  W = np.zeros(len(X[0]))
  eta = 1
  epochs = 100000

  for epoch in range(1, epochs):
    for i, x in enumerate(X):
      if(Y[i]*np.dot(X[i], W)) < 1:
        W = W + eta*((X[i] * Y[i]) + (-2 * (1/epoch) * W))
      else:
        W = W + eta*(-2 * (1/epoch) * W)
  
  return W
```

```python
W = sum_sgd(X, y)
W
```

```python
for i in X:
  val = np.dot(i, W)
  print(np.sign(val))
  # if val < 0:
  #   print(-1)
  # else:
  #   print(1)
```
