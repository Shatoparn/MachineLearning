```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def med(x):
  list.sort(x)
  print("Sorted:",x)
  if(len(x)%2 == 1):
    print("Median:",x[(len(x)//2)])
  else:
    print("Median:",(x[(len(x)//2)-1]+x[(len(x)//2)])/2)
```

```python
a = [4,40,5,1,20,6,7,3,2,21,22,54,10]
med(a)
b=[1,4,3,2]
med(b)
c = [-2,-4,-1,-3]
med(c)
```

```python
def med_mat(l):
  x = [d for row in l for d in row]

  list.sort(x)
  print("Sorted:",x)
  if(len(x)%2 == 1):
    print("Median:",x[(len(x)//2)])
  else:
    print("Median:",(x[(len(x)//2)-1]+x[(len(x)//2)])/2)
```

```python
mat = [[80,40,230],[30,210,70],[11,50,60]]
med_mat(mat)
```

```python
def plot_hist(x):
  y = list(set(x))
  c = []
  for i in y:
    c.append(x.count(i))
  z = [str(i) for i in y]
  ind = c.index(max(c))
  print("Mode :", y[ind])
  plt.bar(z,c,color = ['#000000','r', 'm','#fabf15','#fabfff'])
```

```python
x = [1,1,1,1,4,4,7,7,7,3,2]
plot_hist(x)
```

```python
def N(x,u,s):
  y = (1/math.sqrt(2*3.14*s**2))*np.exp(-((x-u)**2/(2*s**2)))
  return y
```

```python
n = np.linspace(-1,1,100)
n_val = N(n,0.0,0.1)
```

```python
plt.plot(n,n_val,'r')
```

```python
from matplotlib import cm
x=np.linspace(-5,5,100)
y=np.linspace(-5,5,101)
xx,yy=np.meshgrid(x,y)
zz=np.sqrt(xx**2+yy**2)
R=np.sin(zz)
xx.shape,yy.shape,zz.shape

fig, ax=plt.subplots(subplot_kw={"projection":"3d"})
ax.plot_surface(xx,yy,R,cmap=cm.coolwarm,linewidth=0,antialiased=False)
```

```python
def cov(x):
  m=x.mean(axis=0)
  d=x-m
  print('D matrix :',d)
  s=np.dot(d.T,d)
  plt.scatter(x[:,0],x[:,1])
  return s
x=np.array([[1,2],[3,3],[5,4],[6,7]])
print(cov(x))
print('X :',x)
#print(np.shape(x)[1])
#print(np.shape(x)[0])
```
