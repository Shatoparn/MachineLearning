```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
rng = np.random.RandomState(10)
x = rng.rand(10, 1)
y = 6 + 4 * x + 3 * x**2 + 7 * x**3
```

```python
def pre(x):
  x2 = x**2
  x3 = x**3
  m, n = np.shape(x)
  z = np.ones([m, 1])
  x = np.concatenate((z, x, x2, x3), axis = 1)
  x = x/np.max(x, axis = 0)
  return x
```

```python
def grad_desc(x,y,epoch,alpha):
  cost = []
  x = pre(x)
  m = np.shape(x)[0]
  print("x"+ str(x))
  n = np.shape(x)[1]
  w = np.random.rand(n, 1)
  for i in range(epoch):
    h = np.dot(x,w)
    mse = (1/2 * m) * sum((h - y)**2)
    d = h - y
    cost.append(mse)
    w = w - (alpha/m) * (np.dot(x.T,d))
  
  return w, h, cost
```

```python
w, h, cost = grad_desc(x, y, 200, 0.01)
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('center')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

plt.plot(y)
plt.plot(h)
```
