```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def gradient_descent(x,y,alpha=0.01,epoch=30):
  m = np.shape(x)[0]
  n = np.shape(x)[1]
  x = np.concatenate((np.ones((m,1)),x),axis = 1)
  w = np.random.randn(n+1,)
  cost_history_list = []
  for currant_iteration in np.arange(epoch):
    y_estimated = x.dot(w)
    error = y_estimated - y
    cost = (1/2*m)*np.sum(error**2)
    gradient = (1/m)*x.T.dot(error)
    w = w - alpha * gradient
    if currant_iteration % 10 == 0:
      print(f"cost:{cost}\t iteration: {currant_iteration}")
    cost_history_list.append(cost)
  return w, cost_history_list, y_estimated
```

```python
rng = np.random.RandomState(10)
x = 10*rng.rand(24,1)
y = 0.9+np.dot(x,[2.2])
```

```python
weight, cost_history_list, h = gradient_descent(x,y,0.01,50)

plt.plot(np.arange(len(cost_history_list)), cost_history_list)
plt.xlabel("Number of iteration epoch")
plt.ylabel("Cost function J(Theta)")
plt.title("Gradient descent")
plt.show()
```

```python
def plot_regression_line(x, y, b):
    plt.scatter(x, y, '.r')
    y_pred = b[0] + b[1]*x
    plt.plot(x, y_pred, color = "g")
    plt.xlabel('x')
    plt.ylabel('y')
    plt.show()
```

```python
plot_regression_line(x, y, weight)
```
