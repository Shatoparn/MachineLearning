```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.reshape(np.linspace(1,20,20),[20,1])
m = np.shape(x)[0]
y = 2.5*x + 17
x = np.concatenate((np.ones([m,1]),x),axis = 1)
print("Shape of x:",np.shape(x))
#print("X:",x)
X_Tx = np.dot(x.T, x)
print("Shape of X transpose X:",np.shape(X_Tx))
#print("X transpose:",x_T)
X_Ty = np.dot(x.T,y)
print("Shape of X transpose Y:",np.shape(X_Ty))
```

```python
inv = np.linalg.inv(X_Tx)
print(inv)
```

```python
z = np.dot(inv,X_Ty)
```

```python
z
```

```python
h = np.dot(x,z)
plt.plot(h)
plt.plot(y,'.r')
```

```python
def pre(x):
  x2 = x**2
  x3 = x**3
  m, n = np.shape(x)
  z = np.ones([m, 1])
  x = np.concatenate((z, x, x2, x3), axis = 1)
  return x
```

```python
x = np.array([3.0,7.0,20.0,50.0,60.0,65.0])
xx = x.copy()
y = np.array([20.0,50.0,80.0,80.0,50.0,20.0])

x = np.reshape(x,[6,1])
y = np.reshape(y,[6,1])
x = pre(x)
#print(x)
X_Tx = np.dot(x.T, x)
print("Shape of X transpose X:",np.shape(X_Tx))
X_Ty = np.dot(x.T,y)
print("Shape of X transpose Y:",np.shape(X_Ty))
inv = np.linalg.inv(X_Tx)
z = np.dot(inv,X_Ty)
print(z)
print(xx)
```

```python
import numpy as np
X = (np.array([np.linspace(3,65,100)])).T
#print(X)
X = pre(X)

h = np.dot(X,z)
plt.plot(h)
plt.scatter(xx,y)
```

```python
from matplotlib import pyplot as plt
import numpy as np
x = np.array([3.0,7.0,20.0,50.0,60.0,65.0])
xx = x.copy()
y = np.array([20.0,50.0,80.0,80.0,50.0,20.0])
z = np.polyfit(x,y,3)
h = np.poly1d(z)
lin = np.linspace(3,65,100)

plt.scatter(x, y)
plt.plot(lin,h(lin))
plt.show()
```
