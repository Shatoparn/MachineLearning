```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles
from mpl_toolkits.mplot3d import Axes3D

from sklearn import svm
```

```python
X, Y = make_circles(n_samples = 500, noise = 0.02)
plt.scatter(X[:, 0], X[:, 1], c = Y, marker = '.', cmap = 'jet')
plt.show()
```

```python
X1 = X[:, 0].reshape((-1, 1))
X2 = X[:, 1].reshape((-1, 1))
X3 = (X1**2 + X2**2)

X = np.hstack((X, X3))
```

```python
fig = plt.figure()
axes = fig.add_subplot(111, projection = '3d')
axes.scatter(X1, X2, X3, c = Y, depthshade = True, cmap = 'jet')
```

```python
svc = svm.SVC(kernel = 'linear')
svc.fit(X, Y)
w = svc.coef_
b = svc.intercept_
```

```python
w.shape
```

```python
x1 = X[:, 0].reshape((-1, 1))
x2 = X[:, 1].reshape((-1, 1))

x1, x2 = np.meshgrid(x1, x2)
x3 = -(w[0][0]*x1 + w[0][1]*x2 + b) / w[0][2]
```

```python
fig = plt.figure(figsize = (16, 10))
axes1 = fig.gca(projection = '3d')
axes1.scatter(X1, X2, X1**2 + X2**2, c = Y, depthshade = True, cmap = 'jet')
axes1.plot_surface(x1, x2, x3, alpha = 0.005, cmap = 'jet')
# axes1.view_init(10, 9)
plt.show()
```
