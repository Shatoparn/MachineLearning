```python
mux = np.mean(X1)
stdx = np.std(X1)
muy = np.mean(X2)
stdy = np.std(X2)
```

```python
print(mux,muy)
```

```python
print(stdx,stdy)
```

```python
def N(x,u,s):
  y = (1/math.sqrt(2*3.14*s**2))*np.exp(-((x-u)**2/(2*s**2)))
  return y
```

```python
x = np.linspace(-(mux+stdx*5),mux+stdx*5,1000)
y = np.linspace(-(muy+stdy*5),muy+stdy*5,1000)
X = N(x,mux,stdx)
Y = N(y,muy,stdy)
```

```python
plt.plot(x,X)
plt.plot(y,Y)
plt.plot(y,X*Y)
plt.xlim(-20,20)
# plt.ylim(-0.2,1.4)
```

```python
coeX = X - mux
coeY = Y - muy
covX = np.dot(coeX.T,coeX)
covY = np.dot(coeY.T,coeY)
print(covX, covY)
```
